```python
import requests
import json
import pandas as pd
from datetime import datetime
```

```python
results = []

for i in range(1,1001):
  if i % 10 == 0:
    resp = requests.get(f"https://kitsu.io/api/edge/anime?filter[categories]=adventure&sort=popularityRank&page[limit]=10&page[offset]={i}")
    results.extend(resp.json()['data'])
```

```python
data = []
for i,res in enumerate(results) :
  item = {}
  item['id'] = res['id']
  # try :
  #   item["title"] = res['attributes']['titles']['en']
  # except:
  #   item["title"] = res['attributes']['titles']['en_jp']
  item["title"] = res["attributes"]["canonicalTitle"]
  item['synopsis'] = res['attributes']['synopsis']
  data.append(item)
```

```python
def get_main_character(question,context,index):
  data = {
      'question':question,
      'context':context
  }
  header = {
      'accept':'application/json',
      'Content-Type':'application/json'
  }


  resp = requests.post(f'http://{HOST}/',data=json.dumps(data))
  answer = resp.json()["result"]["answer"]
  print(f"index number {index}")
  return answer
```

```python
def get_async_tasks(session,url,data_list):
  tasks = []
  header = {
      'accept':'application/json',
      'Content-Type':'application/json'
  }
  for data in data_list:
    data_json = json.dumps(data)
    tasks.append(session.post(url=url,headers=header,data=data_json))
  return tasks

async def async_request(df:pd.DataFrame):
  start = datetime.now()
  results = []

  url = f"http://{HOST}/"
  num = 1
  data_list = []
  for index,row in df.iterrows():
    data = {
      'question':row["question"],
      'context':row["synopsis"]
    }
    data_list.append(data)
    if num % 1000 == 0 :
        print(f"num of req = {num}")
        async with aiohttp.ClientSession() as session:
          task_submit = get_async_tasks(session,url,data_list)
          response = await asyncio.gather(*task_submit)
          for i,resp in enumerate(response) :
            res = await resp.content.read()
            data = res.decode("UTF-8")
            data_dict = json.loads(data)
            answer = data_dict["result"]["answer"]
            results.append(answer)
            num += 1
            data_list=[]
    else :
        num +=1
        continue
  df["answer"] = pd.Series(results)
  duration = datetime.now() - start
  print(f"time taken {duration.total_seconds()} seconds or {duration.total_seconds()/60} minutes")
  return df
```

### 2. Asynchronous Request

```python
async_df = await async_request(df2)
```
